```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#q1(a)
d = pd.read_csv('company_sales_data.csv')
plt.plot(d['month_number'],d['total_profit'])
plt.title('month profits')
plt.xlabel('month number')
plt.ylabel('total profits')
plt.show()
```

```python
# q1(b)
from matplotlib import style
plt.plot(d['month_number'],d['total_units'],color="r",linestyle="--",linewidth=3,marker="o",label="units line")
plt.title('month profits')
plt.xlabel('month number')
plt.ylabel('sold units number')
plt.legend(loc=4)
```

```python
#q1(c)
plt.plot(d['month_number'],d['toothpaste'],color="b",linestyle="--",linewidth=4,marker="o",markersize=6,label="toothpaste")
plt.plot(d['month_number'],d['facecream'],color="g",linestyle="--",linewidth=4,marker="o",markersize=3,label="facecream")
plt.plot(d['month_number'],d['facewash'],color="c",linestyle="--",linewidth=4,marker="o",markersize=6,label="facewash")
plt.plot(d['month_number'],d['bathingsoap'],color="r",linestyle="--",linewidth=4,marker="o",markersize=6,label="bathingsoap")
plt.plot(d['month_number'],d['shampoo'],color="m",linestyle="--",linewidth=4,marker="o",markersize=6,label="shampoo")
plt.plot(d['month_number'],d['moisturizer'],color="y",linestyle="--",linewidth=4,marker="o",markersize=2,label="moisturizer")
plt.title('sales')
plt.xlabel('month_number')
plt.ylabel('units sold')
plt.legend(loc=4)
plt.xticks(d['month_number'])
plt.yticks()
```

```python
#q1(d)
x=d['month_number']
y=d['toothpaste']

plt.scatter(x,y)
plt.title('toothpaste sales')
xlabel('month_number')
ylabel('toothpaste')
```

```python

#q1(e)
plt.figure(figsize=(16,9))

class_month = np.arange(len(d['month_number']))

width = 0.3

plt.bar(class_month, d['facewash'], width , color = "b",
        label =" facewash")

plt.bar(class_month + width, d['facecream'], width , color = "r",
        label =" facewash")

plt.title('comparison',size=13)
plt.xlabel('month',size=13)
plt.ylabel('units',size=13)
plt.legend(loc=4)
plt.xticks(d['month_number'])
```

```python
#q1(f)
plt.figure(figsize=(16,9))

plt.bar(class_month,d['bathingsoap'],width,color="r",label='bathing soap')
plt.title('bathing soap sales')
plt.xlabel('month')
plt.ylabel('sales')
plt.legend(loc=4)
```

```python
a=d['facecream'].sum()
b=d['facewash'].sum()
c=d['toothpaste'].sum()
e=d['bathingsoap'].sum()
f=d['shampoo'].sum()
g=d['moisturizer'].sum()

class_product = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 
           'shampoo','moisturizer']
class_mean = [a,b,c,e,f,g]
explode = [0,0.03,0.1,0.2,0.3,0.4]

plt.pie(class_mean, labels = class_product,explode=explode,autopct="%0.2f%%")
plt.show()
```

```python
#q1(j)
 

month = d['month_number']
 

a = d['facewash']
 
b = d['facecream']

c = d['toothpaste']

e = d['bathingsoap']

f = d['shampoo']

g = d['moisturizer']
 
# Stackplot with X, Y, colors value
plt.stackplot(month, a , b , c , e , f , g ,
              colors =['r', 'c','k','m','b','g'])
 
plt.xlabel('month')
plt.ylabel('No of units')
 
# Title of Graph
plt.title('Representation of month and units')
 
# Displaying Graph
plt.show()
```
